```python
# Library imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV

# Generation of random data
np.random.seed(0)
X = np.sort(20 * np.random.rand(200,1), axis=0)
y = np.sin(X)[:, 0]
y_noise = np.sin(X)[:, 0] + np.random.normal(size=X.shape[0],scale=0.20)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.scatter(X, y.ravel())
plt.show()
```

```python
plt.xlabel("Feature")
plt.ylabel("Target")
plt.scatter(X, y_noise.ravel())
plt.show()
```

```python
clf = KNeighborsRegressor()
params = {"n_neighbors": range(1, 20)}
clf_gs = GridSearchCV(clf, params, cv=3)
clf_gs.fit(X, y_noise)
print('The best estimator is: {}'.format(clf_gs.best_estimator_))
X_axis = np.linspace(X.min(),X.max(), 100).reshape(100, 1)
y_axis_pred = clf_gs.best_estimator_.predict(X_axis)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.scatter(X, y_noise.ravel())
plt.plot(X_axis, y_axis_pred, c="red")
plt.show()
```
