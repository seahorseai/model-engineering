```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function J(theta) = (theta - 3)^2
def cost_function(theta):
    return (theta - 3) ** 2

# Compute the derivative (gradient) of the cost function
def gradient(theta):
    return 2 * (theta - 3)

# Gradient Descent Function
def gradient_descent(initial_theta, learning_rate=0.1, iterations=50):
    theta = initial_theta
    theta_history = [theta]  # Store theta values for visualization

    for _ in range(iterations):
        theta = theta - learning_rate * gradient(theta)  # Update theta
        theta_history.append(theta)

    return theta, theta_history

# Initialize theta randomly
initial_theta = np.random.uniform(-5, 5)
learning_rate = 0.1
iterations = 50

# Run gradient descent
final_theta, theta_history = gradient_descent(initial_theta, learning_rate, iterations)

# Print results
print(f"Initial theta: {initial_theta:.4f}")
print(f"Final optimized theta: {final_theta:.4f}")

# Plotting the cost function and gradient descent steps
theta_vals = np.linspace(-2, 8, 100)
cost_vals = cost_function(theta_vals)

plt.figure(figsize=(8,5))
plt.plot(theta_vals, cost_vals, label="Cost Function", color='blue')
plt.scatter(theta_history, [cost_function(t) for t in theta_history], color='red', label="Gradient Descent Steps")
plt.xlabel("Theta")
plt.ylabel("Cost J(Theta)")
plt.title("Gradient Descent Optimization")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the cost function J(theta) = (theta - 3)^2
def cost_function(theta):
    return (theta - 3) ** 2

# Compute the gradient (derivative) of the cost function
def gradient(theta):
    return 2 * (theta - 3)  # Return as a scalar or 1D array

# Initial guess for theta
initial_theta = np.random.uniform(-5, 5)

# Perform optimization using scipy's minimize function
result = minimize(cost_function, initial_theta, jac=gradient, method='BFGS')

# Extract optimized theta
optimized_theta = result.x[0]

# Print results
print(f"Initial theta: {initial_theta:.4f}")
print(f"Optimized theta: {optimized_theta:.4f}")
print(f"Number of iterations: {result.nit}")

# Plot the cost function and optimization steps
theta_vals = np.linspace(-2, 8, 100)
cost_vals = cost_function(theta_vals)

plt.figure(figsize=(8,5))
plt.plot(theta_vals, cost_vals, label="Cost Function", color='blue')
plt.scatter(optimized_theta, cost_function(optimized_theta), color='red', marker='o', s=100, label="Optimized Theta")
plt.xlabel("Theta")
plt.ylabel("Cost J(Theta)")
plt.title("Gradient Descent Optimization using SciPy")
plt.legend()
plt.show()
```
